```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

### Dataset Operations

- Read
- Combine
- Modify

```python
df_amd = pd.read_csv('training_dataset/AMD (1980 -11.07.2023).csv')
df_asus = pd.read_csv('training_dataset/ASUS (2000 - 11.07.2023).csv')
df_intel = pd.read_csv('training_dataset/INTEL (1980 - 11.07.2023).csv')
df_nvidia = pd.read_csv('training_dataset/NVIDIA (1999 -11.07.2023).csv')
```

```python
# add extra columns for brand name
df_amd['Brand'] = 'AMD'
df_asus['Brand'] = 'ASUS'
df_intel['Brand'] = 'INTEL'
df_nvidia['Brand'] = 'NVIDIA'
```

```python
# add extra columns for brand name
df_amd['Date'] = pd.to_datetime(df_amd['Date'])
df_asus['Date'] = pd.to_datetime(df_asus['Date'])
df_intel['Date'] = pd.to_datetime(df_intel['Date'])
df_nvidia['Date'] =pd.to_datetime(df_nvidia['Date']) 
```

```python
# combine all the datasets 

df_combined = pd.concat([df_amd, df_asus, df_intel, df_nvidia], ignore_index=True)

df_combined.describe()
```

```python
# convert date
df_combined['Date'] = pd.to_datetime(df_combined['Date'])
```

## Plotting Graphs

### Line Graph

Shows the relationship between two variables, usually time on the x-axis and the corresponding value on the y-axis.

```python
plt.figure(figsize=(10,6))
plt.plot(df_nvidia['Date'], df_nvidia['Close'], label='NVIDIA')
plt.plot(df_asus['Date'], df_asus['Close'], label='ASUS')
plt.plot(df_amd['Date'], df_amd['Close'], label='AMD')
plt.plot(df_intel['Date'], df_intel['Close'], label='Intel')
plt.xlabel('Date')
plt.ylabel('Stock Price (in USD)')
plt.legend()

# Set x-axis to show only years
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Rotate the x-axis labels for better readability
plt.xticks(rotation=65)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot NVIDIA stock prices
axes[0, 0].plot(df_nvidia['Date'], df_nvidia['Close'], color='blue')
axes[0, 0].set_title('NVIDIA Stock Prices')
axes[0, 0].set_xlabel('Date')
axes[0, 0].set_ylabel('Closing Price')
axes[0, 0].xaxis.set_major_locator(mdates.YearLocator())
axes[0, 0].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
axes[0, 0].tick_params(axis='x', rotation=75)

# Plot AMD stock prices
axes[0, 1].plot(df_amd['Date'], df_amd['Close'], color='red')
axes[0, 1].set_title('AMD Stock Prices')
axes[0, 1].set_xlabel('Date')
axes[0, 1].set_ylabel('Closing Price')
axes[0, 1].xaxis.set_major_locator(mdates.YearLocator())
axes[0, 1].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
axes[0, 1].tick_params(axis='x', rotation=75)

# Plot Intel stock prices
axes[1, 0].plot(df_intel['Date'], df_intel['Close'], color='green')
axes[1, 0].set_title('Intel Stock Prices')
axes[1, 0].set_xlabel('Date')
axes[1, 0].set_ylabel('Closing Price')
axes[1, 0].xaxis.set_major_locator(mdates.YearLocator())
axes[1, 0].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
axes[1, 0].tick_params(axis='x', rotation=75)

# Plot ASUS stock prices
axes[1, 1].plot(df_asus['Date'], df_asus['Close'], color='orange')
axes[1, 1].set_title('ASUS Stock Prices')
axes[1, 1].set_xlabel('Date')
axes[1, 1].set_ylabel('Closing Price')
axes[1, 1].xaxis.set_major_locator(mdates.YearLocator())
axes[1, 1].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
axes[1, 1].tick_params(axis='x', rotation=75)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
